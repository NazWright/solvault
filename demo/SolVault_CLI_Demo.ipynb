# SolVault CLI Demo

This notebook demonstrates the SolVault CLI backup functionality.

## Setup

First, ensure you have the dependencies installed:
```bash
npm install
```

## Run Backup Command

Example using a well-known DeGods holder address:
```bash
npm run backup <WALLET_ADDRESS> ./demo-backups
```

````javascript
// Example: backup NFTs for a wallet
// Replace with actual wallet address
const walletAddress = 'YOUR_WALLET_ADDRESS_HERE';
console.log(`Running backup for: ${walletAddress}`);
````

## Expected Output

The backup command will:
1. Fetch all NFTs for the specified wallet
2. Download metadata and images
3. Save to `./demo-backups/<mint-address>/`
4. Display a summary of backed up items
